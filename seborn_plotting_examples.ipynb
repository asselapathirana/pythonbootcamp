```python
import numpy as np
import pandas as pd
import seaborn as sns

data = pd.read_pickle("./data/KNMI_20190815_hourly.zip.pickle", compression="zip")
```

```python
datas = data.sample(n=365*10)
```

```python
# Lets explore the relationship between Pressure and Relative humidity
# but first let's give proper names for our data
tmp = datas.rename(columns={"P":"Pressure (hPa)", "U": "Relative Humidity (%)"})
plot=sns.jointplot(x="Pressure (hPa)", y="Relative Humidity (%)", data=tmp, kind='kde')
```

```python
# Let's compare Pressure, Relative Humidity, Tempreature and total rainfall with each 
tmp = datas[["P","U","T", "RH","S"]]
sns.pairplot(tmp, vars=["P","U","T", "RH"],hue='S', size=2.5);
```

```python
# Let's plot the same graphs with hourly rainfall > 1 mm
datas = data.loc[data['RH'] > 1.0].sample(n=365*10)
# Let's compare Pressure, Relative Humidity, Tempreature and total rainfall with each 
tmp = datas[["P","U","T", "RH","S"]]
sns.pairplot(tmp, vars=["P","U","T", "RH"],hue='S', size=2.5);
```

```python
# is there a relationship between amount of rainfall and pressure 
# Let's plot the same graphs with hourly rainfall > 1 mm
datas = data.loc[data['RH'] > 0.1].sample(n=365*10)
g=sns.jointplot("P", "RH", data=datas, kind='reg', scatter_kws={'alpha':0.1});
```

```python
# Or pressure and relative humidity
datas = data.sample(n=365*10)
g=sns.jointplot("P", "U", data=datas, kind='reg', scatter_kws={'alpha':0.1});
```
